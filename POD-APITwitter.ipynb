El método get_recent_tweets_count da el número de tweets que concuerden con la búsqueda en la última semana, y cada una de estas búsquedas devuelve los tweets que contengan todas las palabras de la query, o bien el hashtag correspondiente.

```python
queries = ["(amazon aws) OR #aws", "(google cloud) OR #gcp OR #googlecloud", "(microsoft azure) OR #azure", "(ibm cloud) OR #ibmcloud"]
resultados = {}

for query in queries:
    recentTweets = client.get_recent_tweets_count(query)
    resultados[query] = round(recentTweets.meta["total_tweet_count"] / 7)
```
